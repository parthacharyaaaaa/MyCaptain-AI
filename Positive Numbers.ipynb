```python
#Input: list1 = [12, -7, 5, 64, -14] Output: 12, 5, 64 Input: list2 = [12, 14, -95, 3] Output: [12, 14, 3]
list1 = [12, -7, 5, 64, -14]

for x in list1:
    if x<0:
        list1.remove(x)
        
print (list1)

list2 = [12, 14, -95, 3]
for x in list2:
    if x<0:
        list2.remove(x)
        
print (list2)

#Hence, both output lists contain only the positive elements

 
```

```python
# Python program to print positive Numbers in given range
  
start, end = -4, 19
  
# iterating each number in list
for num in range(start, end + 1):
      
    # checking condition
    if num >= 0:
        print(num, end = " ")
```
