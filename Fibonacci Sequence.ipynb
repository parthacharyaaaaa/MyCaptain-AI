```python
#Printing Fibonacci sequence

x=0
y=1

n = int(input("Enter number of elements "))

if n == 1:
    print (x)
elif n<=0:
    print ("Number of elements must be greater than 0");
else: 
    print (x)
    print (y)
    
    for i in range (0,n):
        z = x+y
        x=y
        y=z
        print (z)
```
